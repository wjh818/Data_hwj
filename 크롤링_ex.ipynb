## 특정 페이지의 소스 코드를 파일로 저장

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('http://paullab.co.kr/stock.html')
response.encoding = 'utf-8'
html = response.text

f = open('test.html', 'w', encoding = 'utf-8')
f.write(html)
f.close()
```
