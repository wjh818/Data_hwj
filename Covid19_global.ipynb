```python
import pandas as pd

df = pd.read_csv('country_wise_latest.csv')
```

```python
df
```

```python
df["Country/Region"].value_counts()
```

```python
deaths_top10 = df.sort_values(by='Deaths', ascending=False).head(10)
```

```python
deaths_top10
```

```python
deaths_top10 = deaths_top10[['Country/Region', 'Deaths']]
deaths_top10
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.barplot(x='Country/Region', y='Deaths', label='deaths', data=deaths_top10)
plt.xticks(rotation=70)
plt.title("Deaths TOP10")
plt.legend()
plt.show()
```
