```python
titles = soup.find_all('div',{"class": "ellipsis rank01"})
singers = soup.find_all('div',{"class": "ellipsis rank02"})
albums = soup.find_all('div',{"class": "ellipsis rank03"})
```

```python
title = []
singer = []
album = []
```

```python
for i in titles:
    title.append(i.find('a').text)
```

```python
for i in singers:
    singer.append(i.find('span', {"class": "checkEllipsis"}).text)
```

```python
for i in albums:
    album.append(i.find('a').text)
```
