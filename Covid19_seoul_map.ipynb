```python
import folium
import pandas as pd
```

```python
geo_json = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
```

```python
df = pd.read_csv('PatientInfo.csv', usecols=[0,1,4,5,6])
df
```

```python
m = folium.Map(
        location=[37.566345, 126.977893],
        titles='Stamen Terrain')
m
```

```python
is_venezuela = df['province'] == 'Seoul'

df = df[is_venezuela]

row_cut = df['city'] != 'etc'
df = df[row_cut]
df
```

```python
df['city'].value_counts()
```

```python
city_mapping = {'Gwanak-gu':'관악구',
                'Guro-gu':'구로구', 
                'Gangnam-gu':'강남구',          
                'Gangseo-gu':'강서구',          
                'Yangcheon-gu':'양천구',        
                'Dobong-gu':'도봉구',           
                'Yeongdeungpo-gu':'영등포구',     
                'Seocho-gu':'서초구',           
                'Songpa-gu':'송파구',           
                'Dongjak-gu':'동작구',          
                'Eunpyeong-gu':'은평구',        
                'Yongsan-gu':'용산구',
                'Seongdong-gu':'성동구',        
                'Nowon-gu':'노원구',            
                'Dongdaemun-gu':'동대문구',       
                'Mapo-gu':'마포구',             
                'Seodaemun-gu':'서대문구',        
                'Gangdong-gu':'강동구',         
                'Jungnang-gu':'중랑구',         
                'Seongbuk-gu':'성북구',         
                'Geumcheon-gu':'금천구',       
                'Gangbuk-gu':'강북구',          
                'Jongno-gu':'종로구',           
                'Gwangjin-gu':'광진구',         
                'Jung-gu':'중구'             }

df['guname'] = df['city'].map(city_mapping)
```

```python
df.drop('province', axis=1, inplace=True)
```

```python
data_v = df['guname'].value_counts().tolist()
a = df['guname'].tolist()
data_n1 = set(a)
data_n2 = list(data_n1)
df_c = pd.DataFrame([i for i in zip(data_n2, data_v)], columns=['name', 'value'])
df_c
```

```python
from folium import Choropleth

folium.Choropleth(
    geo_data = geo_json,
    name='choropleth',
    data = df_c,
    columns=['name', 'value'],
    key_on='feature.properties.name',
    fill_color='PuBu',
    fill_opacity=0.85, 
    line_opacity=0.2,
).add_to(m)
```
